```python
import numpy as np
import matplotlib.pyplot as plt
import flopy
import os
```

```python
nreal = 50
```

```python
def plot_curve(name, **kwargs):
    fpth = os.path.join("output", f"{name}.mpend")
    e = flopy.utils.EndpointFile(fpth)

    cur2 = []
    time = range(1, 5000, 5)
    for t in time:
        cur2.append(((e.get_alldata().time/3600/24).astype(int) < t).sum())

    plt.plot(time, cur2, **kwargs)
```

```python
for ireal in range(nreal):
    plot_curve(f"EROS{ireal}", color="blue", alpha=0.2)
    plot_curve(f"SIS_{ireal}", color="red", alpha=0.2)
    
plot_curve(f"ref", color="black", alpha=1)
```

```python
import seaborn as sns
```

```python
for ireal in range(nreal):
    name = f"EROS{ireal}"
    fpth = os.path.join("output", f"{name}.mpend")
    e = flopy.utils.EndpointFile(fpth)

    times_data = e.get_alldata().time/3600/24

    # plot density
    sns.kdeplot(times_data, label="EROS1", color="blue", alpha=0.5, linewidth=0.5)

name = f"ref"
fpth = os.path.join("output", f"{name}.mpend")
e = flopy.utils.EndpointFile(fpth)

times_data = e.get_alldata().time/3600/24

# plot density
sns.kdeplot(times_data, color="black", alpha=1, linewidth=0.5)

plt.xlim(0, 1000)
```
